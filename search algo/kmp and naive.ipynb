```python

import numpy as np
from scipy.interpolate import make_interp_spline
import matplotlib.pyplot as plt

# data to be plotted
x = np.array([10,100,1000,5000, 10000 , 100000])
yn = np.array([ 415500,749252,2088007 , 3114543 ,5085031, 28194846 ])
yk = np.array([ 438494, 461180, 761415,2118831,3860863,24745520 ])

# plotting
plt.title("KMP algorithm ")
plt.xlabel("size of text")
plt.ylabel("runtime in nanoseconds")
plt.plot(x, yk,  color ="green" , label = "KMP")
plt.plot(x, yn , color = "red" , label ="naive")
plt.legend()
plt.show()
```
